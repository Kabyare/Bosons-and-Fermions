```python
# import the useful libraries package 
import numpy as np  
import matplotlib.pyplot as plt  
 

### Initialize the constants ###

# Initialize the energy range  
E = np.linspace(-0.5,0.5,10001) 
#electric charge
e = 1.6e-19 
#Boltzmann constant(joule per kelvin) 
k = 1.38e-23  
# chemical potential (\mu) of the substance is zero
u = 0   


# Define function that take two variable (T,a)   
#for Maxwell-Boltxmann distribution a=0, for Bose-Einstein a=-1  
# and for Fermi-Dirac a =+1  

def Fn(T,a):
    result = 1/((np.exp(((E-u)*e)/(k*T)))+a)
    return result  

##  Using plt to display graph for M-B scenario ###
plt.subplot(2,2,1)  
plt.plot(E,Fn(200,0),label='T = 200 K')  
plt.plot(E,Fn(400,0),label='T = 400 K')  
plt.plot(E,Fn(600,0),label='T = 600 K')  
plt.plot(E,Fn(800,0),label='T = 800 K') 
plt.plot(E,Fn(1000,0),label='T = 1000 K')
plt.plot(E,Fn(1200,0),label='T = 1200 K') 
#plt.figure(figsize =(6,8))

plt.grid()  
plt.ylim(0,1) 
plt.ylim(0,3)
plt.xlabel('Energy(eV)',color = 'black',size =15)  
plt.ylabel('f(E)',color = 'black',size =15)  
plt.legend(loc='best',prop={'size':12})  
plt.title("Maxwell-Boltzmann distribution for $\mu =0 $",color ='red',size = 15)  
plt.show()

```

```python
###  Using plt to display graph for B-E scenario ###
plt.subplot(2,2,2)  
plt.plot(E,Fn(200,-1),label='T = 200 K')  
plt.plot(E,Fn(400,-1),label='T = 400 K')  
plt.plot(E,Fn(600,-1),label='T = 600 K')  
plt.plot(E,Fn(800,-1),label='T = 800 K')  
plt.plot(E,Fn(1000,-1),label='T = 1000 K')
plt.plot(E,Fn(1200,-1),label='T = 1200 K')
plt.grid()  
plt.xlim(0,2)  
plt.ylim(0,4)  
plt.xlabel('Energy(eV)',color = 'black',size =15)  
plt.ylabel('f(E)',color = 'black',size =15)  
plt.legend(loc='best',prop={'size':12})        
plt.title("Bose-Einstein distribution for $\mu =0 $",color ='red',size = 15)
plt.show()
```

```python
###  Using plt to display graph for F-D scenario ###
plt.subplot(2,2,3)  
plt.plot(E,Fn(200,+1),label='T = 200 K')  
plt.plot(E,Fn(400,+1),label='T = 400 K')  
plt.plot(E,Fn(600,+1),label='T = 600 K')  
plt.plot(E,Fn(800,+1),label='T = 800 K') 
plt.plot(E,Fn(1000,+1),label='T = 1000 K') 
plt.plot(E,Fn(1200,+1),label='T = 1200 K') 
plt.legend(loc='best',prop={'size':12}) 
plt.grid()  
plt.ylim(0,1)  
plt.xlabel('Energy(eV)',color = 'black',size =15)  
plt.ylabel('f(E)',color = 'black',size =15) 
plt.title("Fermi-Dirac distribution for $\mu =0 $",color ='red',size = 15)  
plt.show()

```

```python
###  Using plt to display graph comparison of three scenario ###
plt.subplot(2,2,4)  
plt.plot(E,Fn(1200,0),label='M-B distribution')  
plt.plot(E,Fn(1200,-1),label='B-E distribution')  
plt.plot(E,Fn(1200,+1),label='F-D distribution')  
plt.legend(loc='best',prop={'size':12})  
plt.ylim(0,4)  
plt.xlabel('Energy(eV)',color = 'black',size =15)  
plt.ylabel('f(E)',color = 'black',size =15)  
plt.title("Temperature  at 1200 K",color ='red',size = 15)


### Sub plot Adjusting ######  

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,
                    top=0.9,wspace=0.4,hspace=0.4)  
plt.grid()
plt.show()  
```

```python
def MB_speed(v,m,T):
    """ Maxwell-Boltzmann speed distribution for speeds """
    kB = 1.38e-23
    return (m/(2*np.pi*kB*T))**1.5 * 4*np.pi * v**2 * np.exp(-m*v**2/(2*kB*T))

fig = plt.figure()
ax = fig.add_subplot(111)

v = np.arange(0,800,1)
amu = 1.66e-27
mass = 85*amu

for T in [100,200,300,400]:
    fv = MB_speed(v,mass,T)
    ax.plot(v,fv,label='T='+str(T)+' K',lw=2)

ax.legend(loc=0)
ax.set_xlabel('$v$ (m/s)')
ax.set_ylabel('PDF, $f_v(v)$')
plt.grid()
plt.draw()
```

```python
from scipy.special import erf

def MB_CDF(v,m,T):
    """ Cumulative Distribution function of
    the Maxwell-Boltzmann speed distribution """
    kB = 1.38e-23
    a = np.sqrt(kB*T/m)
    return erf(v/(np.sqrt(2)*a)) - np.sqrt(2/np.pi)* v* np.exp(-v**2/(2*a**2))/a


fig = plt.figure()
ax = fig.add_subplot(111)

v = np.arange(0,800,1)
amu = 1.66e-27
mass = 85*amu

for T in [100,200,300,400]:
    fv = MB_CDF(v,mass,T)
    ax.plot(v,fv,label='T='+str(T)+' K',lw=2)

ax.legend(loc=0)
ax.set_xlabel('$v$ (m/s)')
ax.set_ylabel('CDF, $C_v(v)$')
plt.grid()
plt.draw()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# Defining the constants
h = 6.626e-34 # Planck constant
k = 1.38e-23 # Boltzman constant
c = 3e+8 # Speed of light in vacuum

# Computing arrays for wavelengths and frequencies
L = np.arange(1e-9, 4e-6, 1e-9) # Lamda
f = np.arange(1e+13, 150e+13, 1e+13) # Nu 

# Defining a function for calculating energy density as a function of wavelength
def planck_lamba(L, T):
    # Breaking down the formula
    a = (8 * math.pi * h * c)/(L ** 5)
    b = (h * c)/(L * k * T)
    E = np.exp(b) - 1
    U = a/E
    return U

# Defining a function for calculating energy density as a function of frequency
def planck_nu(f, T):
    # Breaking down the formula
    a = (8 * math.pi * h * f ** 3)/(c ** 3)
    b = (h * f)/(k * T)
    E = np.exp(b) - 1
    U = a/E
    return U

# Calculation of energy desnsities as functions of wavelengths at different value of temperatures
UL3500 = planck_lamba(L, 3500)# T = 3500 K
UL3500m = np.argmax(UL3500)
UL4000 = planck_lamba(L, 4000)# T = 4000 K
UL4000m = np.argmax(UL4000)
UL4500 = planck_lamba(L, 4500)# T = 4500 K
UL4500m = np.argmax(UL4500)
UL5000 = planck_lamba(L, 5000)# T = 5000 K
UL5000m = np.argmax(UL5000)
UL5500 = planck_lamba(L, 5500)# T = 5500 K
UL5500m = np.argmax(UL5500)

# Ploting above calculations
plt.figure(1)	# Energy density as a function of wavelength
UL3500p= [UL3500m]
UL4000p= [UL4000m]
UL4500p= [UL4500m]
UL5000p= [UL5000m]
UL5500p= [UL5500m]
plt.annotate('2000K', xy=(745, 102), xytext=(745, 102), fontstyle = "italic")
plt.plot(L*1e+9, UL3500*1e-3, marker = "o", markevery = UL3500p)
plt.annotate('3000K', xy=(640, 188), xytext=(640, 188), fontstyle = "italic")
plt.plot(L*1e+9, UL4000*1e-3, marker = "o", markevery = UL4000p)
plt.annotate('4000K', xy=(555, 328), xytext=(555, 328), fontstyle = "italic")
plt.plot(L*1e+9, UL4500*1e-3, marker = "o", markevery = UL4500p)
plt.annotate('5000K', xy=(490, 546), xytext=(490, 546), fontstyle = "italic")
plt.plot(L*1e+9, UL5000*1e-3, marker = "o", markevery = UL5000p)
plt.annotate('6000K', xy=(435, 872), xytext=(435, 872), fontstyle = "italic")
plt.plot(L*1e+9, UL5500*1e-3, marker = "o", markevery = UL5500p)
plt.ylim(0,920)
plt.xlim(100, 2000)
plt.xlabel(r"$\lambda$ / 10$^{-9}$ [m]", fontsize = 12, fontweight = "bold")
plt.ylabel(r"U($\lambda$,T) / 10$^{3}$ [J/m$^{4}$]", fontsize = 12, fontweight = "bold")
plt.title("Energy density as a function of wavelength", fontsize = 14, color = "green", fontweight = "bold", fontstyle = "italic")
plt.grid()
plt.show()

# Calculation of energy desnsities as functions of frequencies at different value of temperatures
UF3500 = planck_nu(f, 3500)# T = 3500 K
UF3500m = np.argmax(UF3500)
UF4000 = planck_nu(f, 4000)# T = 4000 K
UF4000m = np.argmax(UF4000)
UF4500 = planck_nu(f, 4500)# T = 4500 K
UF4500m = np.argmax(UF4500)
UF5000 = planck_nu(f, 5000)# T = 5000 K
UF5000m = np.argmax(UF5000)
UF5500 = planck_nu(f, 5500)# T = 5500 K
UF5500m = np.argmax(UF5500)

# Ploting above calculations
plt.figure(2)# Energy density as a function of frequency
UF3500p= [UF3500m]
UF4000p= [UF4000m]
UF4500p= [UF4500m]
UF5000p= [UF5000m]
UF5500p= [UF5500m]
plt.annotate('2000K', xy=(15, 36), xytext=(15, 36), fontstyle = "italic")
plt.plot(f*1e-13, UF3500*1e+17, marker = "o", markevery = UF3500p)
plt.annotate('3000K', xy=(23, 101.5), xytext=(23, 101.5), fontstyle = "italic")
plt.plot(f*1e-13, UF4000*1e+17, marker = "o", markevery = UF4000p)
plt.annotate('4000K', xy=(20, 74.5), xytext=(20, 74.5), fontstyle = "italic")
plt.plot(f*1e-13, UF4500*1e+17, marker = "o", markevery = UF4500p)
plt.annotate('5000K', xy=(18.5, 53), xytext=(18.5, 53), fontstyle = "italic")
plt.plot(f*1e-13, UF5000*1e+17, marker = "o", markevery = UF5000p)
plt.annotate('6000K', xy=(26, 134), xytext=(26, 134), fontstyle = "italic")
plt.plot(f*1e-13, UF5500*1e+17, marker = "o", markevery = UF5500p)
plt.ylim(0, 142)
plt.xlim(0, 140)
plt.xlabel(r"$\nu$ / 10$^{13}$ [Hz]", fontsize = 12, fontweight = "bold")
plt.ylabel(r"U($\nu$,T) / 10$^{-17}$ [Js/m$^{3}$]", fontsize = 12, fontweight = "bold")
plt.title("Energy density as a function of frequency", fontsize = 14, color = "green", fontweight = "bold", fontstyle = "italic")
plt.grid()
plt.show()

# Calcuations which lead to Ultraviolet Catostrophe
f_ct = (8 * math.pi * f ** 2 * k * 5000) / c ** 3# Energy density per unit frequency
L_ct = (8 * math.pi * k * 5000) / L ** 4# Energy density per unit wavelength

#Ploting based on the above two results
plt.figure(3)# Classical Vs Qunatum Result for U(L, 5000 K)
plt.plot(L*1e+9, UL5000*1e-3)
plt.plot(L*1e+9, L_ct*1e-3, "--", color = "m")
plt.annotate('Ultraviolet', xy=(950,730), xytext=(500, 700), fontweight = "bold", fontstyle = "italic")
plt.annotate('Catostrophe', xy=(950,700), xytext=(450, 670), fontweight = "bold", fontstyle = "italic")
plt.annotate('At T = 5000K', xy=(2400,370), xytext=(2400,370), fontsize = 15)
plt.ylim(0,750)
plt.xlim(100, 4000)
plt.legend(['Planck Law | Qunatum', 'Rayleigh-Jeans Law | Classical']  , loc = 'best')
plt.xlabel(r"$\lambda$ / 10$^{-9}$ [m]", fontsize = 12, fontweight = "bold")
plt.ylabel(r"U($\lambda$,T) / 10$^{3}$ [J/m$^{4}$]", fontsize = 12, fontweight = "bold")
plt.title(r"Classical Vs Qunatum Result for U($\lambda$, 5000 K)", fontsize = 14, color = "r", fontweight = "bold", fontstyle = "italic")
plt.grid()
plt.show()

plt.figure(4)# Classical Vs Quantum Result for U(f, 5000 K)
plt.plot(f*1e-13, UF5000*1e+17, label = "Planck Law | Qunatum")
plt.plot(f*1e-13, f_ct*1e+17, "--", label = "Rayleigh-Jeans Law | Classical", color = "m")
plt.annotate('Ultraviolet', xy=(20,133), xytext=(18.5,133), fontweight = "bold", fontstyle = "italic")
plt.annotate('Catostrophe', xy=(18,128), xytext=(16.5,128), fontweight = "bold", fontstyle = "italic")
plt.annotate('At T = 5000K', xy=(85,65), xytext=(85,65), fontsize = 15)
plt.ylim(0, 140)
plt.xlim(0, 140)
plt.legend()
plt.xlabel(r"$\nu$ / 10$^{13}$ [Hz]", fontsize = 12, fontweight = "bold")
plt.ylabel(r"U($\nu$,T) / 10$^{-17}$ [Js/m$^{3}$]", fontsize = 12, fontweight = "bold")
plt.title(r"Classical Vs Qunatum Result for U($\nu$, 5000 K)", fontsize = 14, color = "y", fontweight = "bold", fontstyle = "italic")
plt.grid()
plt.show()
```

```python
 
import numpy as np  
from scipy.constants import h, c, k, pi  
import matplotlib.pyplot as plt
newparams = {'axes.labelsize': 10, 'axes.linewidth': 1.5, 'savefig.dpi':  
      1000,  
     'lines.linewidth': 2, 'figure.figsize': (18, 12),    
     'legend.frameon': True,  
      'legend.handlelength': 0.7}  

plt.rcParams.update(newparams)  

L0 = np.arange(0.1,30,0.005) #Wavelength in micro m
L = L0*1e-6 #wavelength in m


def planck_lamda(L,T):  
     a = (8 * pi * h * c) / L **5  
     b = (h * c)/ (L * k * T)  
     c1 = np.exp(b) - 1  
     d = a / c1  
     return d
plt.suptitle("""Plancks law for Black Body radiation & compare it with Raleigh-Jeans
             Law at high & low temperature""",  
          size = 14,color='green',fontstyle ="italic")      
    
R_Ht = (8* pi *k *1100)/L**4 #Rayleigh's law at High temperature  
R_Lt = (8*pi*k*200)/L**4 #Rayleigh's law at Low temperature

T500 = planck_lamda(L , 500)  
T900 = planck_lamda(L , 900)
T1100 = planck_lamda(L , 1100)
T1300 = planck_lamda(L , 1300)


""" ##### Sub plot creation ##### """
plt.subplot(2,2,(1,2))
plt.plot(L, T500,label='T=500 K')  
plt.plot(L, T900 ,label='T=900 K')
plt.plot(L, T1100 ,label='T=1100 K')
plt.plot(L, T1300 ,label='T=1300 K')
plt.legend(loc="best" ,prop={'size':12})
plt.xlabel(r"$\lambda$ ")  
plt.ylabel(r"U($\lambda $,T )")
plt.title("Planck Law of Radiation")
plt.ylim(0,20)
plt.xlim(0,0.00002)
plt.grid()

plt.subplot(2,2,3)
plt.plot(L, (planck_lamda(L,200)),label='Planck Law')
plt.plot(L, R_Lt , "--" , label="Rayleigh-Jeans Law")
plt.legend(title="Comparison at low Temperature",loc="best" ,prop={'size':12})
plt.xlabel(r"$\lambda$ ")  
plt.ylabel(r"U($\lambda $,T )")
plt.title("T=200 K (For Low Temperature)")
plt.ylim(0,0.5)
plt.xlim(0,0.00003)
plt.grid()
  
plt.subplot(2,2,4)
plt.plot(L, T1100 ,label='Planck Law')
plt.plot(L, R_Ht , "--" , label="Rayleigh-Jeans Law")
plt.legend(title="Comparison at high Temperature",loc="best" ,prop={'size':12})
plt.xlabel(r"$\lambda$ ")  
plt.ylabel(r"U($\lambda $,T )")
plt.title("T=1100 K (For High Temperature)")
plt.ylim(0,15)


"""###### Sub plot Adjusting ######"""
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.7, 
                    top=0.7, 
                    wspace=0.4, 
                    hspace=0.4)

plt.grid()
plt.show()          
```
